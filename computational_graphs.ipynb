```python
p = cg.Variable("p")
q = cg.Variable("q")
k = cg.Variable(3)
expression = cg.sqrt(p + p + q) / k + cg.ln(p * (p * q)) + k
expression.graph()
```

```python
x1 = cg.Variable("x1")
x2 = cg.Variable("x2")
f = cg.ln(x1) + x1 * x2 - cg.sin(x2)
f.graph()
```
