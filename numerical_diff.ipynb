```python
import numpy as np
import matplotlib.pyplot as plt
```

# Numerical Differentiation (via Finite Differences)

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    sig = sigmoid(x)
    return sig * (1 - sig)

x = np.linspace(-5, 5)
plt.plot(x, sigmoid(x), label="$\sigma(x)$")
plt.plot(x, sigmoid_derivative(x), label="$\sigma '(x)$")
plt.legend()
```

Recall the limit definition of the derivative: $f'(x) = \lim_{h \to 0}\frac{f(x+h) - f(x)}{h}$. It follows that we can approximate the derivative if we fix a small $h$. While this definition is perfect in theory, in practice dividing by a small $h$ on a computer leads to floating point error and dividing by a large $h$ gives a bad approximation.

```python
inputs = [-2, 0, 4]
h_list = [10 ** i for i in range(-10, 11)]
for x in inputs:
    errors = []
    d_true = sigmoid_derivative(x)
    for h in h_list:
        d_pred = (sigmoid(x + h) - sigmoid(x)) / h
        errors.append(np.abs(d_true - d_pred))
    plt.plot(h_list, errors, label=f"$x={x}$")
plt.legend()
plt.ylabel("Absolute Error")
plt.yscale('log')
plt.xlabel("$h$")
plt.xscale('log')
```
